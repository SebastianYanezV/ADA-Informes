# 1. Descripción del problema: Problema de Ordenamiento
**Entrada:** Lista o arreglo de "n" números $[a_1, a_2, ..., a_n]$

**Salida:** La lista o arreglo de la entrada, pero ahora ordenada: $[a_1',a_2',...,a_n']$, de forma que $a_1'\leq a_2' \leq... \leq a_n'$.

El problema que se quiere resolver en este caso es el problema de ordenamiento, el cuál consiste en ordenar de forma eficiente una secuencia de "n" números en desorden. El algoritmo que se usará para resolver este problema será el *Merge Sort*, el cúal fue desarrollado en 1945 en la búsqueda de encontrar un algoritmo eficiente que resuelva este famoso problema.

# 2. Merge Sort

El código que se muestra a continuación pertenece al algoritmo *Merge Sort*, el cúal contiene las funciones *MergeSort* y *Merge*.


## 2.1 Código

```python
llamadas = 0
comparaciones = 0

def Merge(sub1, sub2, verbose = False):
  #Se crea el arreglo en el que se almacenarán los elementos de los dos sub arreglos de forma ordenada
  largo = len(sub1) + len(sub2)
  new = [0] * (largo)

  #Se inicializan los sub índices de cada arreglo
  i = 0 ; j = 0 ; k = 0

  #Se van traspasando los elementos de ambos sub arreglos al nuevo arreglo ordenado
  while (len(sub1) > i and len(sub2) > j):
    if (sub1[i] < sub2[j]):
      new[k] = sub1[i]
      i += 1
    else:
      new[k] = sub2[j]
      j += 1
    k += 1
    global comparaciones
    comparaciones += 1

  #En el momento en el que un sub arreglo se queda vacío, los datos restantes del otro se agregan directamente en el arreglo ordenado
  while (i < len(sub1)):
    new[k] = sub1[i]
    i += 1
    k += 1
    comparaciones += 1

  while (j < len(sub2)):
    new[k] = sub2[j]
    j += 1
    k += 1
    comparaciones += 1
  
  return new

def mergeSort(A, verbose = False):
  #Si el largo del arreglo es igual a 1, se retorna el mismo arreglo
  if len(A) == 1: return A

  if verbose == True: print("Al inicio de la pasada: ", A)

  #Esta variable se utiliza para dividir cada arreglo por la mitad
  medio = int(len(A)/2)

  sub1 = mergeSort(A[:medio])
  if verbose == True: print("Sub arreglo izquierdo: ", sub1)
  global llamadas
  llamadas += 1

  sub2 = mergeSort(A[medio:])
  if verbose == True: print("Sub arreglo derecho: ", sub2)
  llamadas += 1

  if verbose == True: print("Arreglos que entran al Merge: ", sub1, "y", sub2)
  if verbose == True: print("Al salir del Merge: ", Merge(sub1, sub2))
  return Merge(sub1, sub2)

# Ejemplo
A = [5, 1, 13, 4, 8, 54, 10]
print("Arreglo de entrada: ", A)
A = mergeSort(A)
print("Arreglo de salida: ", A)
print("LLamadas recursivas de Merge Sort: ", llamadas)
print("Comparaciones realizadas por Merge: ", comparaciones)
```

## 2.2 Descripción del algoritmo

Antes de empezar a ordenar, el algoritmo recibe una lista o arreglo de "n" números en desorden, los que deberá ordenar a continuación.

En primer lugar, trabaja la función *MergeSort*.

1. El algoritmo se encarga de sub dividirse en dos sub arreglos de largo (n/2). Esto lo realiza de forma iterativa hasta quedar con varios sub arreglos de largo 1.
2. Luego se encarga de ordenar cada sub arreglo de forma recursiva (llamándose a si mismo) gracias a la función *Merge*.

La función *Merge* recibe ambos sub arreglos como parámetros y funciona de la siguiente manera.

1. Se crea un nuevo arreglo en el que se almacenarán todos los elementos de ambos sub arreglos de forma ordenada.
2. Se compara el primer elemento del primer sub arreglo con el primer elemento del segundo sub arreglo y se añade al menor de ellos en el nuevo arreglo ordenado. Luego, tanto el sub índice del sub arreglo que "cedió" su elemento al arreglo ordenado, como el sub índice del nuevo arreglo aumentan en una posición. 
3. El paso anterior se repite hasta llegar al final de alguno de los dos sub arreglos.
4. Cuando un sub arreglo traspasa todos sus elementos al arreglo ordenado, los elementos restantes del otro sub arreglo se traspasan automáticamente al arreglo ordenado.

Luego, todos los pasos anteriormente descritos se realizan recursivamente hasta terminar de ordenar completamente el arreglo original. Finalmente, al terminar el proceso anteriormente descrito, el algoritmo retorna el arreglo de entrada ordenado de menor a mayor.

## 2.3 Ejemplo

Se ingresa el siguiente arreglo:

A = [2, 5, 8, 1]

Luego se divide en dos sub arreglos distintos:

B = [2, 5] 

C = [8, 1]

Después, se ordenan ambos sub arreglos de forma recursiva, resultando en:

B = [2, 5]

C = [1, 8]

Para terminar, la función *Merge* se encarga de combinar ambos sub arreglos en un nuevo arreglo:

D = [1, 2, 5, 8]

Finalmente se retorna el arreglo ordenado.

## 2.4 Ejecución del algoritmo paso a paso

```python
#import random
#A = random.sample(range(1, 100), 4)
A = [66, 3, 27, 50]
print("Arreglo original: ", A, "\n")
A = mergeSort(A, verbose = True)
print("\nArreglo ordenado: ", A)
```

Como se puede ver en el ejemplo anterior, el arreglo original se divide en dos sub arreglos, [66, 3] y [27, 50]. En primer lugar, subdivide nuevamente el primer sub arreglo mencionado, resultando en los sub arreglos [66] y [3] y luego los combina de forma ordenada gracias a la función *Merge*, dando como resultado [3, 66]. En segundo lugar, realiza los mismos pasos con el sub arreglo [27, 50]. Finalmente, cuando los dos sub arreglos principales se encuentran ordenados, los ingresa una vez más a la función *Merge* para terminar el ordenamiento del arreglo, el cual finalmente es retornado como [3, 27, 50, 66].

# 4. Correctitud

## Teorema: Correctitud de la función Merge

El algoritmo *Merge* genera un arreglo: $[a_1',a_2',...,a_n']$, con los mismos elementos de los subarreglos de entrada ordenados de menor a mayor, es decir, $a_1'\leq a_2' \leq... \leq a_n'$.

## Prueba del teorema

A continuación se probará la siguiente propiedad de la función *Merge*, la cuál corresponde a su **propiedad invariante de bucle**:

"Luego de cada comparación entre ambos sub arreglos ordenados (sub1 y sub2), el nuevo arreglo (new), el arreglo en el que se colocan los elementos ordenados, contiene los mismos elementos que el arreglo original, pero ahora ordenados de menor a mayor. Además, todos los elementos ingresados en el arreglo ordenado son menores que los que aún no se han ingresado, por lo que el mayor elemento del arreglo ordenado, será menor que el siguiente elemento a ingresar."

**Inicialización**

Antes de la primera iteración del ciclo "while (len(sub1) > i and len(sub2) > j)", el arreglo "new" se encuentra vacío y por consecuencia no contiene ningún elemento de los sub arreglos ordenados (sub1 y sub2), y como los sub índices "i" y "j" son iguales a 0, eso quiere decir que sub1[i] y sub2[j] son los menores elementos de sus respectivos arreglos que no han sido copiados al arreglo ordenado "new".

**Mantención**

Para ver que la propiedad invariante del bucle se mantiene en cada iteración, supongamos que $sub1[i] < sub2[j]$. En este caso, sub1[i] sería el menor elemento que no ha sido copiado en "new". Como "new" contiene los "k" menores elementos, entonces sub1[i] se debería copiar en "new" para que este contenga los "k + 1" menores elementos. Luego de este paso, tanto "k" como "i" aumentan sus valores en 1, por lo que la propiedad invariante de bucle se reinicia para la siguiente iteración.

## Teorema: Correctitud del algoritmo MergeSort

El algoritmo *Merge Sort* genera un arreglo: $[a_1',a_2',...,a_n']$, con los mismos elementos del arreglo de entrada ordenados de menor a mayor, es decir, $a_1'\leq a_2' \leq... \leq a_n'$.

## Prueba del teorema

Para poder probar lo anterior, se debe usar el método de inducción matemática (ya que se está tratando con un algoritmo recursivo), en el que se debe probar un hipótesis P(n). Para esto se necesita probar que tanto el caso base como el paso inductivo son correctos.

**Caso base**

Se toma el caso más sencillo, cuando el arreglo de entrada es de tamaño 1. En este caso, el *Merge Sort* retornará el arreglo ordenado ya que este solo posee un elemento. Se cumple P(1).

**Paso inductivo**

Aquí se debe probar que P(m), con m < n, es cierto (n = tamaño del arreglo original). Cuando se tiene un arreglo de tamaño "n", se crean dos sub arreglos de tamaño n/2 y estos se ordenan gracias a *Merge Sort*. Como estos sub arreglos se ordenan pese a tener un tamaño m < n, se asume que el paso inductivo es correcto.

**Correctitud**

Finalmente, como el caso base y paso inductivo son correctos, y como luego la función *Merge* se encarga de unir y ordenar ambos sub arreglos, se concluye que el teorema de correctitud es verdadero.

## 5.2 Merge Sort v/s Insertion Sort (comparaciones)

A continuación se muestra una gráfica en la que se comparan la cantidad de comparaciones promedio entre los algoritmos *Merge Sort* e *Insertion Sort* para tamaños de problemas con n ∈ [5, 99].

```python
def insertionSort(A, verbose = False):
  largo = len(A) #Se calcula el largo del arreglo
  cont = 0 #Se inicializa el contador de comparaciones

  #En este ciclo se leen los elementos desde el segundo hasta el último
  for i in range(1, largo): 
    key = A[i] 
    
    #En este ciclo se leen los elementos de la izquierda del "key"
    for k in range(i - 1, -1, -1): 
      if (A[k] < key): #Si "key" es mayor a algún elementos de su izquierda, este se coloca justo a su derecha
        A[k+1] = key
        break;

      cont += 1
      A[k+1] = A[k] #Si "key" no es mayor a los elementos de su izquierda, entonces este se va moviendo hacia su izquierda
        
    if (key < A[0]): #Si el primer elemento es mayor que "key", entonces este último se mueve a la primera posición 
      A[0] = key 
      
  return A, cont 
```

```python
x=[n for n in range(5,100)] 
y=[]; 
z=[];

for n in range(5,100):
  a = random.sample(range(1, 100), n)
  b = random.sample(range(1, 100), n)

  comparaciones = 0
  a = mergeSort(a)

  b, cont = insertionSort(b)

  y.append(comparaciones)
  z.append(cont)

plt.plot(x,y)
plt.plot(x,z)
plt.legend(["Merge Sort", "Insertion Sort"])

plt.xlabel('n')
plt.ylabel('Número de comparaciones')
plt.show()
```

### Análisis de resultados

En el gráfico se puede ver claramente que el algoritmo *Insertion Sort* realiza más comparaciones que el *Merge Sort*. Esto ocurre debido a que el tiempo de ejecución del primer algoritmo mencionado en el peor caso es de $O(n^2)$, mientras que el tiempo de ejecución del *Insertion Sort* es de $O(n*logn)$ para cualquier caso.

## 5.2 Merge Sort v/s Insertion Sort (tiempo de ejecución)
