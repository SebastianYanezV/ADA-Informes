# 1. Descripción del problema: Problema de Ordenamiento

**Entrada:** Lista o arreglo de n números $[a_1,a_2,...,a_n]$

**Salida:** La lista o arreglo de la entrada, pero ahora ordenada: $[a_1',a_2',...,a_n']$, de forma que $a_1'\leq a_2' \leq... \leq a_n'$.

El problema que se busca resolver al utilizar el algoritmo "Insertion Sort" es el clásico problema de ordenamiento, en el cuál se busca ordenar una secuencia de "n" números desordenados, esto de la forma más eficiente posible.

Cabe recalcar que, si bien este algoritmo de ordenamiento sirve para poder ordenar una secuencia de "n" números, este no es el algoritmo más eficiente para poder resolver este problema.

# 2. Insertion Sort

El código que se muestra a continuación pertenece al algoritmo *Insertion Sort*.

## 2.1 Código

```python
def insertionSort(A, verbose = False):
  largo = len(A) #Se calcula el largo del arreglo
  cont = 0 #Se inicializa el contador de comparaciones

  if verbose == True: print("Arreglo original:", A)

  #En este ciclo se leen los elementos desde el segundo hasta el último
  for i in range(1, largo): 
    key = A[i] 

    if verbose == True: print("\nPasada", i)

    if verbose == True: print("\nNúmero leído (key):", key)

    if verbose == True: print("Al inicio de la pasada:", A)
    
    #En este ciclo se leen los elementos de la izquierda del "key"
    for k in range(i - 1, -1, -1): 
      if (A[k] < key): #Si "key" es mayor a algún elementos de su izquierda, este se coloca justo a su derecha
        A[k+1] = key
        if verbose == True: print("Se encuentra un número menor a la izquierda del key:", A)
        break;

      cont += 1
      A[k+1] = A[k] #Si "key" no es mayor a los elementos de su izquierda, entonces este se va moviendo hacia su izquierda
      if verbose == True: print("El key se va moviendo hacia su izquierda:", A)
        
    if (key < A[0]): #Si el primer elemento es mayor que "key", entonces este último se mueve a la primera posición 
      A[0] = key 
      if verbose == True: print("El key se inserta en la primera posición:", A)
    
    if verbose == True: print("Al final de la pasada:", A)
  
  if verbose == True: print("\nArreglo ordenado:", A)

  mediaRetrocesos = cont/(largo-1) #La media de los retrocesos se calcula como el total de comparaciones dividido en en largo del arreglo menos 1
      
  return A, cont, mediaRetrocesos 
  
#Ejemplo
A = [4, 3, 2, 1]
print("Entrada:", A)
A, cont, media = insertionSort(A)
print("Salida:", A)
print("Comparaciones:", cont)
print("Media Retrocesos:", media)
```

## 2.2 Descripción del algoritmo
Antes de empezar el ordenamiento, el algoritmo recibe una lista o arreglo de "n" números en desorden, los que deberá ordenar a continuación.

1. El algoritmo va leyendo los números del conjunto uno a uno, cada vez que lee uno, se encarga de revisar si este está ordenado o no, esto lo logra comparando el número leído con todos los números que se encuentran a su izquierda.

2. Si el número leído es menor al que está a su izquierda, este debe intercambiar su posición. Esto se realiza hasta que se encuentre un número mayor al leído o hasta que este mismo llegue a la primera posición del arreglo.

3. Al terminar de leer toda la lista o arreglo, el algoritmo retorna el conjunto de números, esta vez ordenados de menor a mayor.

## 2.3 Ejemplo

Se ingresa el siguiente arreglo:

A = [3, 2, 1]

En la primera vuelta se intercambia el 2 con el 3 y queda así: 

A = [2, 3, 1]

Luego se avanza y se intercambia la posición del 1 con el 3:

A = [2, 1, 3]

Finalmente, en la segunda vuelta, se intercambia el 1 con el 2: 

A = [1, 2, 3]

Finalmente el algoritmo está ordenado.

## 2.4 Ejecución del algoritmo paso a paso

A continuación se muestra la ejecución paso a paso gracias a la opción verbose = true.

```python
import random
A = random.sample(range(1, 100), 6)
A, cont, media = insertionSort(A, verbose = True)
print("Media Retrocesos:", media)
```

Como se puede apreciar en el ejemplo anterior, cada vez que se lee un nuevo número, todos los números del arreglo, desde la posición 0 hasta la posición "i" en donde está el número leído, se ordenan de forma ascendente. Esto ocurre debido a que cada vez que se lee un nuevo número, este se empieza a mover hacia su izquierda hasta que encuentre un número menor que él o hasta llegar al inicio del arreglo.

# 3. Tiempo de Ejecución

## Teorema: Tiempo de Ejecución

El algoritmo *Insertion Sort* tiene un tiempo de ejecución de O($n^2$) en el **peor caso**.

## Prueba del teorema

En el caso del *Insertion Sort*, su tiempo de ejecución se calcula en base a la cantidad de comparaciones que este realiza al momento de ordenar. Gracias al ejemplo presente en la sección del código (2.1), se puede observar que el **peor caso** ocurre cuándo el arreglo está ordenado al revés de como se quiere ordenar (en este caso, de mayor a menor en vez de menor a mayor). 

## Ejemplo

```python
A = [4, 3, 2, 1]
A, cont, media= insertionSort(A, verbose = True)
print("Comparaciones:", cont)
print("Media Retrocesos:", media)
```

En el ejemplo anterior, en la primera pasada hizo 1 comparación, en la segunda hizo 2 y en la tercera hizo 3.

Además, al examinar más casos se puede deducir que sabiendo la cantidad de datos y comparaciones realizadas, se puede calcular las comparaciones del siguiente "nivel". Por ejemplo, si se sabe que con 6 elementos se realizan 15 comparaciones, entonces podemos saber que con 7 elementos se realizarán 21 comparaciones, ya que 6 + 15 = 21. O también, sabiendo lo anterior, podemos calcular que con 8 elementos se realizarán 28 comparaciones, ya que 7 + 21 = 28.

Entonces, de forma generalizada, este algoritmo debería hacer un total de $n(n-1)/2$ comparaciones. Por lo que se puede confirmar que el tiempo de ejecución del algoritmo en el **peor caso** es O($n^2$).

# 4. Correctitud

## Teorema: Correctitud

El algoritmo *Insertion Sort* genera un arreglo: $[a_1',a_2',...,a_n']$, con los mismos elementos del arreglo de entrada ordenados de menor a mayor, es decir, $a_1'\leq a_2' \leq... \leq a_n'$.

## Prueba del teorema

Gracias a los ejemplos anteriores, se puede observar que el sub arreglo que va desde la posición 0 hasta la posición "i - 1" en la que está el elemento "key" al inicio de la pasada, contiene todos los elementos originales leídos, pero ahora ordenados.

Por esto, se probará la siguiente propiedad del algoritmo *Insertion Sort*, la cuál corresponde a su **propiedad invariante de bucle**:

"Al comienzo de cada iteración del bucle más grande, el subarreglo A[0..i-1] contiene los mismos elementos que el arreglo original, pero ordenados de menor a mayor". 

**Inicialización**

Al inicio del algoritmo, i = 1 por lo que el sub arreglo [0, i-1] es igual a la posición 0, por lo que inmediatamente se puede saber que este sub arreglo se encuentra ordenado.

**Mantención**

Luego, se asume que cada vez que se itera, el sub arreglo [0, i-1] se encuentra ordenado y mantiene a los elementos del arreglo original. Luego, en cada pasada se toma al elemento leído de la posición "i" (key) y se lo mueve hasta su posición correcta. Por lo tanto, cada vez que este paso se termine de hacer, el sub arreglo [0, i-1] se encontrará ordenado y seguirá teniendo los elementos del arreglo original.

**Correctitud**

Finalmente, como las afirmaciones de la inicialización y mantención son correctas, se puede concluir que la propiedad invariante del bucle es correcta y se mantiene durante todo el algoritmo, por lo que el teorema de correctitud es correcto.

# 5. Experimentos

## 5.1 Cantidad de comparaciones

A continuación se muestra una gráfica en la que se comparan la cantidad de comparaciones entre el **mejor caso** y **peor caso** del algoritmo para tamaños de problemas con n ∈ [5, 19].

```python
import matplotlib.pyplot as plt

x = [n for n in range(5,20)] 
y1 = [n*(n-1)/2 for n in range(5,20)] # peor caso
y2 = [n-n for n in range(5,20)] # mejor caso
y = []; 

for n in range(5,20):
  a = random.sample(range(1, 100), n)
  a, counter, media = insertionSort(a)
  y.append(counter)

plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(["Insertion Sort", "Peor caso teórico", "Mejor caso teórico"])

plt.xlabel('n')
plt.ylabel('Número de comparaciones')
plt.show()
```

### Análisis de resultados

En el gráfico se puede apreciar que en el mejor caso la gráfica se mantiene como una línea recta sobre 0, ya que en este caso el arreglo está ordenado, por lo que no se deben hacer comparaciones.

Por otro lado, se puede ver como la cantidad de comparaciones en el peor caso aumenta progresivamente a medida que aumenta la cantidad de elementos en el arreglo.

## 5.2 Arreglo casi ordenado

A continuación se muestra una gráfica en la que se muestra lo que ocurre al momento de ordenar un arreglo casi ordenado en comparación con otro arreglo en orden aleatorio.

```python
import matplotlib.pyplot as plt

x = [n for n in range(5,20)] 
y2 = [n-1 for n in range(5,20)]
y = []; 

for n in range(5,20):
  a = random.sample(range(1, 100), n)
  a, counter, media = insertionSort(a)
  y.append(counter)

plt.plot(x,y)
plt.plot(x,y2)
plt.legend(["Arreglo aleatorio", "Arreglo casi ordenado"])

plt.xlabel('n')
plt.ylabel('Número de comparaciones')
plt.show()
```

### Análisis de resultados

En el gráfico se puede apreciar claramente que un arreglo casi ordenado necesita realizar menos comparaciones para poder ordenarse, a diferencia de un arreglo que esté en orden aleatorio. Esto debido a que se deben realizar menos intercambios de posición para ordenar el arreglo.

## 5.3 Comparación de tiempos de ejecución (Insertion Sort v/s Bubble Sort)

En los gráficos que se muestran a continuación se visualizan los tiempos de ejecución de los algoritmos *Insertion Sort* y *Bubble Sort* en función del tamaño del problema para arreglos generados aleatoriamente de tamaños entre 5 y 500.

```python
import matplotlib.pyplot as plt
import datetime
from timeit import repeat

x=[]; y=[]

for n in range(5,500):

  a = random.sample(range(1, 1000), n)

  #la función repeat está ejecutando 20 veces insertionSort con el arreglo a y retorna el mejor tiepo obtenido.
  t = repeat(setup="from __main__ import insertionSort", stmt=f"insertionSort({a})", repeat=1, number=10)

  x.append(n)
  y.append(t)


plt.plot(x,y)
plt.legend(["Insertion Sort"])

plt.xlabel('n')
plt.ylabel('time in ms')
plt.show()
```
